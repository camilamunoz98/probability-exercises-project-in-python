## Ejercicio 1 

Dos dados se lanzan una vez y se observa el total obtenido. Usa una simulación para encontrar la probabilidad estimada de que la puntuación sea mayor a 7 o un número par. Una simulación es una repetición del mismo experimento multiples veces para observar su comportamiento:

- Ejecuta el experimento 1000 veces (lanza 2 dados 1000 veces, y suma el número de ambos dados).
- Lleva ua cuenta de los números y las veces que la suma fue mayor a 7 o un número par.
- Divide el número del paso 2 entre el número de iteraciones(1000).

```python
import random

def simular_dados(n_iteraciones = 1000):
    exitos = 0
    for _ in range (n_iteraciones):
        dado1 = random.randint(1,6)
        dado2 = random.randint(1,6)
        suma = dado1 + dado2
        
        if suma > 7 or suma %2 == 0:
            exitos += 1
    probabilidad = exitos/n_iteraciones
    return probabilidad
resultado = simular_dados()
print ('Probabilidad estimada:', resultado*100, '%')

```

## Ejercicio 2

Una caja contiene 10 bolas blancas, 20 bolas rojas y 30 bolas verdes. Si tomamos 5 bolas de la caja reemplazándolas (tomando una bola, anotando el color y luego regresandola a la caja). Queremos saber la probabilidad de:

1. Tomar 3 blancas y 2 rojas.
2. Toas todas del mismo color.

Ejecuta el experimento 1000 veces y calcula las probabilidades que mencionadas.

```python
ball_box = {}

# Crea la cajas con las bolas
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

print(ball_box)
            
```

```python
import random 

def simulacion_experimento (n_iteraciones = 1000):
    tres_blancas_dos_rojas = 0
    mismo_color = 0

    for _ in range(n_iteraciones):

        #Sacar 5 bolas con reemplazo 
        seleccion = [random.choice(ball_box) for _ in range(5)]

        #contar los colores
        blancas = seleccion.count('White')
        rojas = seleccion.count('Red')

        #Ejecutar la condiccion 1: 3 blancas y 2 rojas
        if blancas == 3 and rojas == 2:
            tres_blancas_dos_rojas +=1
        #Ejecutar la condiccion 2: todas del mismo color 
        if len(set(seleccion)) == 1:
            mismo_color +=1
    #Calculo la probabilidad
    p_blancas_rojas = tres_blancas_dos_rojas/n_iteraciones
    p_mismo_color = mismo_color/n_iteraciones

    return p_blancas_rojas, p_mismo_color


p1, p2 = simulacion_experimento()
print("Probabilidad de 3 blancas y 2 rojas:", round(p1*100,2) , '%')
print("Probabilidad de todas del mismo color:", round(p2*100,2), '%')
```
